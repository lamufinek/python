```python
from math import sin, cos, sqrt, atan2, radians
import matplotlib.patches as mpatches
import matplotlib.gridspec as gridspec
import pandas as pd
import datetime
from numpy import *
from datetime import datetime, timedelta
from scipy.stats import invgauss, betaprime
import matplotlib.pyplot as plt
import scipy.stats as st
from sklearn.cluster import KMeans
import re


R = 6373.0
```

```python
df_powerPlants = pd.read_csv("global_power_plant_database.csv")
df_cities = pd.read_excel("worldcities.xlsx")
```

```python
df_citiesUkraine = df_cities[df_cities["iso3"]=="UKR"]
df_citiesUkraine.head()
```

```python
df_powerPlantsUkraine = df_powerPlants[df_powerPlants["country"]=="UKR"]
df_powerPlantsUkraine.head()
```

```python
table_AT = pd.read_html('https://en.wikipedia.org/wiki/Anti-Terrorist_Operation_Zone_(Ukraine)', match='Geographic coordinates')
print(f'Total tables: {len(table_AT)}')
df_Anti_Terrorist_Operation_Zone = table_AT[0]
df_Anti_Terrorist_Operation_Zone.head()
```

```python
df_lenght = len(df_Anti_Terrorist_Operation_Zone)
AT_zone_latitude = []
AT_zone_longitude = []

for i in range(1,df_lenght,2):
    string = df_Anti_Terrorist_Operation_Zone.iloc[[i],[1]]
    string = str(string)

    
    stringTable = re.findall(r'\d+', string)
    s= stringTable[2]+"."+stringTable[3]+stringTable[4]
    AT_zone_latitude.append(float(s))
    

for i in range(2,df_lenght,2):
    string = df_Anti_Terrorist_Operation_Zone.iloc[[i],[1]]
    string = str(string)

    
    stringTable = re.findall(r'\d+', string)
    s= stringTable[2]+"."+stringTable[3]+stringTable[4]
    AT_zone_longitude.append(float(s))
    
northLatitudeWarZone = max(AT_zone_latitude)
westLongitudeWarZone = min(AT_zone_longitude)
print(northLatitudeWarZone)
print(westLongitudeWarZone)
```

```python
df_powerPlantsUkraineWarZone = df_powerPlantsUkraine[(df_powerPlantsUkraine["latitude"]<northLatitudeWarZone) & (df_powerPlantsUkraine["longitude"]>westLongitudeWarZone) ]
df_powerPlantsUkraineWarZone.head()

        
    
```

```python
df_powerPlantsUkraineStableZone = df_powerPlantsUkraine[(df_powerPlantsUkraine["latitude"]>northLatitudeWarZone) | (df_powerPlantsUkraine["longitude"]<westLongitudeWarZone) ]
df_powerPlantsUkraineStableZone.head()
```

```python
deltaLatitudes = []
for i in range(0,len(df_powerPlantsUkraineStableZone)):
    latitude  = df_powerPlantsUkraineStableZone.iloc[[i],[5]].values[0]
    latitudeATCity = AT_zone_latitude[0]
    #latitude  = df_powerPlantsUkraineStableZone.iat[5,10]
    deltaLatitudeWarZone = abs(AT_zone_latitude[0] - latitude)
    for latitudeAT in AT_zone_latitude:
        delta = abs(latitudeAT - latitude)
    #print(delta)
    #print(delta)
        if(delta<deltaLatitudeWarZone):
            deltaLatitudeWarZone = delta[0]
            latitudeATCity = latitudeAT
            
    deltaLatitudes.append(latitudeATCity)
    
    
deltaLongitudes = []
for i in range(0,len(df_powerPlantsUkraineStableZone)):
    longitude  = df_powerPlantsUkraineStableZone.iloc[[i],[6]].values[0]
    logitudeATCity = AT_zone_longitude[0]
    #latitude  = df_powerPlantsUkraineStableZone.iat[5,10]
    deltaLongitudeWarZone = abs(AT_zone_longitude[0] - longitude)
    for logitudeAT in AT_zone_longitude:
        delta = abs(logitudeAT - longitude)
    #print(delta)
    #print(delta)
        if(delta< deltaLongitudeWarZone):
            deltaLongitudeWarZone = delta[0]
            logitudeATCity = logitudeAT
    deltaLongitudes.append(logitudeATCity)
    
    
print(deltaLatitudes)
print(deltaLongitudes) 
```

```python
distances = []
for i in range(len(deltaLatitudes)):
    
    lat1 = radians(deltaLatitudes[i])
    lon1 = radians(deltaLongitudes[i])
    
    latitude  = df_powerPlantsUkraineStableZone.iloc[[i],[5]].values[0]
    lat2 = radians(latitude[0])
    
    longitude  = df_powerPlantsUkraineStableZone.iloc[[i],[6]].values[0]
    lon2 = radians(longitude[0])
    
    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    
    distance = R * c
    
    distances.append(distance)
    
print(distances)
```

```python
#df_powerPlantsUkraineStableZone["Distances_AT_Zone"] = distances
df_powerPlantsUkraineStableZone.insert(loc=8,
         column="Distances_AT_Zone",
         value=distances)
df_powerPlantsUkraineStableZone.head()
```

```python

kmeans = KMeans(n_clusters=3, random_state=0)
#df_powerPlantsUkraineStableZone['cluster'] = kmeans.fit_predict(df_powerPlantsUkraineStableZone[['capacity_mw','Distances_AT_Zone']])# get centroids

df_powerPlantsUkraineStableZone.insert(loc=8,
         column="cluster",
         value=kmeans.fit_predict(df_powerPlantsUkraineStableZone[['estimated_generation_gwh','Distances_AT_Zone']]))


centroids = kmeans.cluster_centers_
cen_x = [i[1] for i in centroids] 
cen_y = [i[0] for i in centroids]
## add to df
df_powerPlantsUkraineStableZone.insert(loc=8,
         column="cen_x",
         value=df_powerPlantsUkraineStableZone.cluster.map({0:cen_x[0], 1:cen_x[1], 2:cen_x[2]}))

df_powerPlantsUkraineStableZone.insert(loc=8,
         column="cen_y",
         value=df_powerPlantsUkraineStableZone.cluster.map({0:cen_y[0], 1:cen_y[1], 2:cen_y[2]}))
#df_powerPlantsUkraineStableZone['cen_x'] = df_powerPlantsUkraineStableZone.cluster.map({0:cen_x[0], 1:cen_x[1], 2:cen_x[2]})
#df_powerPlantsUkraineStableZone['cen_y'] = df_powerPlantsUkraineStableZone.cluster.map({0:cen_y[0], 1:cen_y[1], 2:cen_y[2]})# define and map colors
colors = ['#DF2020', '#81DF20', '#2095DF']
#df_powerPlantsUkraineStableZone['c'] =df_powerPlantsUkraineStableZone.cluster.map({0:colors[0], 1:colors[1], 2:colors[2]})
df_powerPlantsUkraineStableZone.insert(loc=8,
         column="c",
         value=df_powerPlantsUkraineStableZone.cluster.map({0:colors[0], 1:colors[1], 2:colors[2]}))
```
